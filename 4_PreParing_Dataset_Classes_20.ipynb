```python
from keras.preprocessing.image import ImageDataGenerator
from skimage import io
import numpy as np
import os
from PIL import Image
```

## **1. Batik Bali**

```python
import os
import shutil

source = '/content/drive/MyDrive/Capstone Project/Batik_DataSet_V2/batik-bali'
destination = '/content/drive/MyDrive/Capstone Project/Batik_DataSet_V2_Ver1/batik-bali-dataset-V1'

# Gather all files
allfiles = os.listdir(source)

# Iterate over all files to move the .jpg files to the destination folder
for filename in allfiles:
    if filename.endswith('.jpg'):
        src_path = os.path.join(source, filename)
        dst_path = os.path.join(destination, filename)
        shutil.copy(src_path, dst_path)
```

```python
datagen = ImageDataGenerator(
    rotation_range=180, 
    fill_mode='reflect')


image_directory = '/content/drive/MyDrive/Capstone Project/Batik_DataSet_V2/batik-bali/'
SIZE = 512
dataset = []

my_images = os.listdir(image_directory)
for i, image_name in enumerate(my_images):
    if (image_name.split('.')[-1].lower() == 'jpg'):
        image = io.imread(image_directory + image_name)
        image = Image.fromarray(image, 'RGB')
        image = image.resize((SIZE,SIZE))
        dataset.append(np.array(image))

x = np.array(dataset)

i = 0
for batch in datagen.flow(x, batch_size=15,  
                          save_to_dir='/content/drive/MyDrive/Capstone Project/Batik_DataSet_V2/batik-bali/augmented',
                          save_prefix='aug', 
                          save_format='jpg'):
    i += 1
    if i > 8:
        break  # otherwise the generator would loop indefinitely  
```

## **2. Batik Betawi**

```python
import os
import shutil

source = '/content/drive/MyDrive/Capstone Project/Batik_DataSet_V2/batik-betawi'
destination = '/content/drive/MyDrive/Capstone Project/Batik_DataSet_V2_Ver1/batik-betawi-dataset-V1'

# Gather all files
allfiles = os.listdir(source)

# Iterate over all files to move the .jpg files to the destination folder
for filename in allfiles:
    if filename.endswith('.jpg'):
        src_path = os.path.join(source, filename)
        dst_path = os.path.join(destination, filename)
        shutil.copy(src_path, dst_path)
```

```python
datagen = ImageDataGenerator(
    rotation_range=180, 
    fill_mode='reflect'
)

image_directory = '/content/drive/MyDrive/Capstone Project/Batik_DataSet_V2/batik-betawi/'
SIZE = 512
dataset = []

my_images = os.listdir(image_directory)
for i, image_name in enumerate(my_images):
    if image_name.split('.')[-1].lower() == 'jpg':
        try:
            image_path = os.path.join(image_directory, image_name)
            image = Image.open(image_path).convert('RGB')  # Open image in RGB mode
            image = image.resize((SIZE, SIZE))
            dataset.append(image)
        except (IOError, OSError) as e:
            print(f"Error loading image: {image_name}, Error message: {str(e)}")

if len(dataset) == 0:
    print("No valid images found in the directory.")
else:
    processed_dataset = []
    for image in dataset:
        processed_image = np.array(image)
        processed_dataset.append(processed_image)

    x = np.array(processed_dataset)
    i = 0
    for batch in datagen.flow(x, batch_size=10,  
                              save_to_dir='/content/drive/MyDrive/Capstone Project/Batik_DataSet_V2/batik-betawi/augmented',
                              save_prefix='aug', 
                              save_format='jpg'):
        i += 1
        if i > 8:
            break  # otherwise the generator would loop indefinitely 
```
